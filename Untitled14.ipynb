```python
import pandas as pd

# Load train
X_train = pd.read_csv("UCI HAR Dataset/train/X_train.txt", delim_whitespace=True, header=None)
y_train = pd.read_csv("UCI HAR Dataset/train/y_train.txt", delim_whitespace=True, header=None)

# Load test
X_test = pd.read_csv("UCI HAR Dataset/test/X_test.txt", delim_whitespace=True, header=None)
y_test = pd.read_csv("UCI HAR Dataset/test/y_test.txt", delim_whitespace=True, header=None)

# Combine train and test
X = pd.concat([X_train, X_test], axis=0)
y = pd.concat([y_train, y_test], axis=0)

# Rename target column
y.columns = ["Activity"]

# Combine features + target
har_df = pd.concat([X, y], axis=1)

# Map activity labels to names
activity_map = {
    1: "Walking",
    2: "Walking_Upstairs",
    3: "Walking_Downstairs",
    4: "Sitting",
    5: "Standing",
    6: "Laying"
}

har_df["Activity"] = har_df["Activity"].map(activity_map)

# Save to CSV
har_df.to_csv("har_dataset.csv", index=False)

print("Dataset saved as har_dataset.csv")
print("Shape:", har_df.shape)
```

## ANN MODEL

```python
import tensorflow as tf
from tensorflow.keras import models,layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import relu,softmax
from tensorflow.keras.callbacks import EarlyStopping
```

```python
y_train = y_train - 1
y_test = y_test - 1
```

```python
y_train = y_train.values.flatten()
y_test = y_test.values.flatten()
```

```python
model = Sequential([
    layers.Dense(256,activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(128,activation='relu'),
    layers.Dense(6,activation="softmax")


])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=25,validation_data=(X_test,y_test),batch_size=64)
```

```python
train_acc = history.history['accuracy'][-1]
test_acc = history.history['val_accuracy'][-1]
print("Train Accuracy:", train_acc)
print("Test Accuracy:", test_acc)
```
